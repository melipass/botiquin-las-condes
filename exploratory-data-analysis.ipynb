```python
import numpy as np
```

```python
import math
```

```python
rows = []
for lab in sorted(df.index.unique()):
    row = []
    row.append(lab)
    if (type(df['Precio referencial con convenio junio 2021'].loc[lab]) != np.int64):
        row.append(df['Precio referencial con convenio junio 2021'].loc[lab].count())
    else:
        row.append("1")
    row.append(math.trunc(df['Precio referencial sin convenio junio 2021'].loc[lab].mean()))
    row.append(math.trunc(df['Precio referencial con convenio junio 2021'].loc[lab].mean()))
    row.append(math.trunc(df['Precio referencial sin convenio junio 2021'].loc[lab].mean()) - math.trunc(df['Precio referencial con convenio junio 2021'].loc[lab].mean()))
    rows.append(row)
df_mean_per_lab = pd.DataFrame(rows, columns=['Laboratorio', 'Cantidad medicamentos', 'Promedio sin convenio', 'Promedio con convenio', 'Diferencia promedios'])
df_mean_per_lab
```

```python
df_mean_per_lab.sort_values('Diferencia promedios', ascending=False).head(20)
```

```python
for lab in df_mean_per_lab.sort_values('Diferencia promedios', ascending=False).head(20)['Laboratorio']:
    if (type(df.loc[lab]) == pd.core.frame.DataFrame):
        display(df.loc[lab].sort_values('Descuento', ascending=False))
    else:
        display(df.loc[lab].to_frame().T)
```

```python
for lab in df_mean_per_lab.sort_values('Diferencia promedios', ascending=False).head(20)['Laboratorio']:
    if (type(df.loc[lab]) == pd.core.frame.DataFrame):
        display(df.loc[lab].sort_values('Descuento', ascending=False))
    else:
        display(df.loc[lab].to_frame().T)
```
